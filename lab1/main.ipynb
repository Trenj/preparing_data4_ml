3. Создание нового набора из 100 первых записей и нужных полей

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем исходный файл
df = pd.read_csv(
    'JCMB_2006_Jun.csv',
    encoding='latin1'
)

# Смотрим названия столбцов
print(df.columns)
```

Создаём новый набор из нужных полей:

```python
new_df = df[
    [
        'date-time',
        'atmospheric pressure (mBar)',
        'wind speed (m/s)',
        'wind direction (degrees)',
        'surface temperature (C)',
        'relative humidity (%)'
    ]
].head(100)

new_df.to_csv('weather_100_rows.csv', index=False)

new_df.head()
```

Создан новый CSV-файл: weather_100_rows.csv

4. Открыть сохранённый файл и вывести данные

```python
saved_df = pd.read_csv('weather_100_rows.csv')

# Первые 20
print("Первые 20 записей:")
print(saved_df.head(20))

# Последние 20
print("Последние 20 записей:")
print(saved_df.tail(20))

# Все записи
print("Все записи:")
print(saved_df)
```

5. Количество уникальных значений relativehumidity (%)

```python
unique_count = saved_df['relative humidity (%)'].nunique()

print("Количество уникальных значений relative humidity (%):")
print(unique_count)
```

Вывод списка уникальных значений:

```python
print(saved_df['relative humidity (%)'].unique())
```

6. 10 наибольших значений windspeed (m/s) + график

```python
top_10_wind = saved_df['wind speed (m/s)'].nlargest(10)

print("10 наибольших значений wind speed (m/s):")
print(top_10_wind)

# Строим график
top_10_wind.plot()
plt.title("10 наибольших значений wind speed (m/s)")
plt.xlabel("Индекс")
plt.ylabel("Скорость ветра (m/s)")
plt.show()
```
